```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random

def rnd(n):
    x, y = 0, 0

    # Generate the time points [1, 2, 3, ... , n]
    tp = np.arange(n + 1)
    pos = [y]

    direct = ["UP", "DOWN"]
    for i in range(1, n + 1):

        # Randomly select either UP or DOWN
        dt = random.choice(direct)
        
        # Move the object up or down
        if dt == "UP":
            y += 1
        elif dt == "DOWN":
            y -= 1

        # Keep track of the pos
        pos.append(y)

    return tp, pos

rw1 = rnd(1000)
rw2 = rnd(1000)
rw3 = rnd(1000)

plt.plot(rw1[0], rw1[1], 'r-', label="rw1")
plt.plot(rw2[0], rw2[1], 'g-', label="rw2")
plt.plot(rw3[0], rw3[1], 'b-', label="rw3")

plt.show()

ar1 = np.array(rw1)
def make_df(x):
    df = pd.DataFrame(np.array(x).T, columns = ['t', 'y'])
    df.set_index('t', inplace = True)
    return df

df1 = make_df(rw1)
df2 = make_df(rw2)
df3 = make_df(rw3)

#print(df1.index)
dftot = pd.DataFrame(np.array([df1['y'], df2['y'], df3['y']]).T, columns = ['wakler1', 'walker2', 'walker3'], index = df1.index)
display(dftot)
dfcorr = dftot.corr('pearson')
dftot.to_csv('walkers.csv')
dfcorr.to_csv('walkers_corr.csv')
```
